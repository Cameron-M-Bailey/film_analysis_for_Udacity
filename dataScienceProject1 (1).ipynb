## Feature Engineering

In order for the analysis to be effective, we must measure the dollar value on the same scale. Thus the dollar amounts for budget and revenue muse be adjusted for inflation. Our analysis uses Consumer Index Price (CPI) for Index Value



```python
# We need to adjust revenue and budgets for inflation, make a column for the year
# Let's pull in the data from the CPI table
df['year'] = df['release_date'].dt.year
```

```python
# Rename column
cpi = cpi.rename(columns={'Year':'year'})
```

## Function
Since the CPI values were pulled directly from a table on a website, theya re not in the needed format. The fucntion below will trim the excess characters on each CPI value.

```python
# Data was pulled from website, need to reformat
def clean_year(value):
    '''
    returns cleaned CPI year for the inflation conversion rate, if is already in the correct format it will ignore the value.
    
    input:
        value (year) with excess charcters i.e. many of the year dates look like \xa01990
    
    output:
        clean_value with the excess characters removed i.e. \xa01990 will now be 1990
    '''
    value = str(value)
    if '\xa0' in value:
        return value.replace('\xa0','').strip()
    return value
```

```python
# Clean the CPI year columns
cpi['year'] = cpi['year'].apply(clean_year)
cpi['year'] = pd.to_datetime(cpi['year'])
cpi['year']= cpi['year'].dt.year
```

```python
# Join the two dataframes on year
dfWithCPI = pd.merge(df,cpi, on='year', how='left')
```

```python
# CPIs are only reliable from 1913-2024, remove other years
dfWithCPI = dfWithCPI.dropna(subset=['Annual Average CPI'])
```

```python
# Add current CPI for all rows 
dfWithCPI["current_cpi"] = 314.4
```

```python
# Calculate the adjsuted revenues and budgets with inflation 
dfWithCPI['adj_revenue'] = dfWithCPI["revenue"] * (dfWithCPI['current_cpi']/dfWithCPI['Annual Average CPI'])
dfWithCPI['adj_budget'] = dfWithCPI["budget"] * (dfWithCPI['current_cpi']/dfWithCPI['Annual Average CPI'])
```

The adjusted revenue and adjusted budget columns make it so that the dollar amount in 1913 accurately equates to the dollar amount in 2024.

# Section 5a: Evaluate the Results

## Data Analysis

### Question 1: Adjusted for inflation do most movies breakeven or turn a profit?

```python
# Calculate how many movies have broke even or earned more after being adjusted for inflation
dfWithCPI['breakeven'] = np.where(dfWithCPI['adj_revenue'] >= dfWithCPI['adj_budget'], 'Y', 'N')
```

First, let's comapare the revenues and budgets of each movie. We've categorized mvoies into Y and N. Y means their revenue equals or exceeds their initial budget. And we can see that by and large, movie earn back the money spent. 

```python
sns.countplot(x='breakeven',data=dfWithCPI)
```

### Question 2: Are movie runtimes significantly increasing and is it impacting viewer reviews?

Here we are calculating the average movie runtime per year. We can determine from the graph below that indeed movie runtimes have increased over time but in last 10-20 years, the increase are marginal. 

```python
# Aggregate the average runtime per year
average_runtime = dfWithCPI.groupby('year')['runtime'].mean().reset_index()
```

```python
# Movies have slightly been increasing run time over time
plt.figure(figsize=(30,10))
sns.barplot(x='year', y='runtime', data=average_runtime)
```

### Question 3: Is there a relationship between the budget and revenue or movie popularity?

Let's now create a correlation matrix to see hpw stronglt our features interact with each other. 

```python
# Create numeric dataframe for correlation analysis 
numeric_df = dfWithCPI.select_dtypes(include='number')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(numeric_df.corr(),annot=True)
```

With the knowledge that movie times are slightly increasing, we can look at the correlation analysis to determine that as runtimes increase, movie popularity increases as well. Meaning that the assumption that longer runtimes are lower movie enjoyment is not true.

The correlation analyssi also addressing question #3, and we can determine that both adjusted revnenue and budget do indeed increase the popularity of a movie.

# Section 5b: Modeling and Results

## Linear Regression

### Question 4: Is it possible to forecast revenue with the data provided?

```python
# Based on the corrlatrion analysis, we can drop columns with multicollinearity
dfWithCPI.drop(['budget','revenue'],axis=1,inplace=True)
```

```python
# Should also drop columns that don't make sense in forecasting revenue.
dfWithCPI.drop(['vote_average','vote_count','status','year','Annual Average CPI','current_cpi','breakeven'],axis=1, inplace=True)
```

```python
# Year most likely won'e be helpful but month of release could be useful information. let's add that feature
dfWithCPI['month'] = dfWithCPI['release_date'].dt.month
```

```python
dfWithCPI.drop(['release_date'],axis=1, inplace=True)
```

```python
# We have no categorical values, so no need to covnert to dummies
```

```python
# Let's remove the target variable from the input features
y = dfWithCPI['adj_revenue']
X = dfWithCPI[['runtime', 'adj_budget', 'month']]
```

```python
# Standardize data prior to model implementation
scaler = StandardScaler()
scaled_df = scaler.fit_transform(X)
```

```python
# Split the data train and test sets
X_train, X_test, y_train, y_test = train_test_split(scaled_df,y, test_size=0.3, random_state=101)
```

```python
# Initiate Linear Regression
lm = LinearRegression()
```

```python
# Fit to the train data and test on the training data
lm.fit(X_train,y_train)
y_train_pred = lm.predict(X_train)
mean_squared_error(y_train,y_train_pred)
```

```python
# Test on the testing data
y_test_pred = lm.predict(X_test)
mean_squared_error(y_test,y_test_pred)
```

Overall this is a bad model but good starting point. No signs of overfitting, instead underfitting based on results.
More relevant features would need to be added to reliablely predict movie revenue. More examples of features to be added have been included in the full write up on Medium.
